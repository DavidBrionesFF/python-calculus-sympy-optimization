## Graficas de funciones en Sympy

```python
from sympy import *
from sympy.plotting import *
```

$ f(x) = 3x + 4$

```python
x = symbols("x")
```

```python
x
```

```python
funcion = 3 * x + 4
```

```python
funcion
```

```python
plot(funcion)
```

```python
funcion_cuadratica = x ** 2 + funcion
```

```python
funcion_cuadratica
```

```python
plot(funcion_cuadratica, ylim=(-50, 50), xlim=(-50, 50))
```

```python
funcion_cubica = 3 * x ** 3 + (4.5 * funcion_cuadratica)
```

```python
funcion_cubica
```

```python
plot(funcion_cubica, ylim=(-500, 500), xlim=(-500, 500))
```

```python
plot(funcion_cuadratica, ylim=(-100, 100), xlim=(-100, 100))
```

```python
diff(funcion_cuadratica, x)
```

```python
funcion_derivada_cuadratica = diff(funcion_cuadratica, x)
```

```python
funcion_derivada_cuadratica
```

```python
p1 = plot(funcion_cuadratica, (x, -10, 10))
p2 = plot(funcion_derivada_cuadratica, (x, -10, 10))

p1.append(p2[0])
```

```python
p1.show()
```
