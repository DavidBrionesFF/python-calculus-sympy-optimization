$\displaystyle \left(X^{T} X\right)^{-1} X^{T} Y$

```python
import random
 
def gen_data(n, bias, varianza):
    x = []
    y = []
    for i in range(0, n):
        x.append([i, 1])
        
        y.append((i + bias) + random.uniform(0, 1) * varianza)
    return x, y
 
x,y = gen_data(100, 25, 50)


plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot([e[0] for e in x], y, marker="o", linewidth=0)
plt.show()
```

```python
print(f"Tamaño de x = {len(x)}")
print(f"Tamaño de y = {len(y)}")
```
